##  What is Machine Learning?

**Machine Learning (ML)** is a subset of artificial intelligence that focuses on enabling computers to **learn from data** instead of being explicitly programmed.

###  Core Idea:
Rather than writing fixed rules or logic, we feed data to a computer, and it **learns patterns** from it. Once trained, the system can **make decisions or predictions** on new, unseen data.

###  Example:
Imagine giving a model hundreds of photos of characters labeled "Harry," "Hermione," and "Ron." You don’t tell it 'how' to recognize them — instead, it figures out their unique features (like Harry's glasses or Ron's red hair) and learns to classify new photos on its own.

>  Machine Learning = Data + Pattern Recognition → Autonomous Decision Making


##  k-Nearest Neighbors (k-NN) Classification

###  Nearest Neighbor Approach

One simple way to solve a classification task is to assign a data point the label of the **closest known observation**.

>  Example :  
> A new student at Hogwarts is being sorted into a house. If their magical behavior is closest to that of a known **Gryffindor** student, then they are assigned to **Gryffindor**.

This method works well in many cases — **but not always**.

---

###  The Problem With Just One Neighbor

Sometimes, the closest neighbor might be an **outlier**. For example:

- A new student is closest to one **Slytherin**, but most nearby students are **Gryffindor**.
- Assigning them to Slytherin based on just one neighbor would likely be a mistake.

---

###  k-Nearest Neighbors to the Rescue

Instead of just one neighbor, use **k nearest neighbors**, and assign the most common label among them.

>  With **k = 3**, if two of the three closest students are from Gryffindor and one from Slytherin,  
> the new student will be assigned to **Gryffindor**.

This method makes classification more robust by **considering more context**.

## Perceptron Learning:

The **Perceptron algorithm** is a machine learning model used for binary classification tasks. In this example, we'll use the Perceptron to classify whether a character in the Harry Potter universe belongs to **Gryffindor** or **Slytherin**, based on two features: **Courage** and **Ambition**.

### Problem Overview:
We are given data with two features:
- **Courage** (x₁)
- **Ambition** (x₂)

The goal is to predict whether a character belongs to **Gryffindor** or **Slytherin**. We'll represent these as:
- **1**: Gryffindor
- **0**: Slytherin

We aim to learn a linear decision boundary that separates the two houses based on these features.

### Mathematical Model:
The Perceptron model for classification is represented as a **linear equation**:

\[
h(x_1, x_2) = w_0 + w_1x_1 + w_2x_2
\]

Where:
- **\(w_0\)** is the bias term (intercept),
- **\(w_1\)** and **\(w_2\)** are the weights for **Courage** and **Ambition** respectively,
- **\(x_1\)** and **\(x_2\)** are the features (Courage and Ambition).

The output is:
- **Gryffindor (1)** if the output of the hypothesis is greater than or equal to 0.
- **Slytherin (0)** otherwise.

Our goal is to learn the best weight vector **w** (i.e., **w₀, w₁, w₂**) using the Perceptron learning rule.

### Perceptron Learning Rule:
The perceptron learning rule adjusts the weights based on the classification errors. For each data point:

1. **Predicted Output** \( \hat{y} \) is computed as:
   \[
   \hat{y} = \text{sign}(w_0 + w_1x_1 + w_2x_2)
   \]
   Where the **sign()** function returns **1** if the result is positive (Gryffindor), and **0** if negative (Slytherin).

2. **Weight Update Rule**:
   \[
   w_i = w_i + \alpha \cdot (y - \hat{y}) \cdot x_i
   \]
   Where:
   - **\(y\)** is the actual label (1 for Gryffindor, 0 for Slytherin),
   - **\(\hat{y}\)** is the predicted label,
   - **\(\alpha\)** is the learning rate,
   - **\(x_i\)** is the corresponding input feature value (Courage or Ambition).

   This update rule adjusts the weights after each misclassified data point to improve the model’s accuracy.

### Example Dataset:
Consider the following data for three characters in the Harry Potter universe:

|----------------|--------------|---------------|---------|

| Draco Malfoy   | 3            | 8             | Slytherin (0)   |
| Hermione Granger | 9          | 4             | Gryffindor (1) |

We want the Perceptron algorithm to learn from this dataset and predict the house for new characters.

###  Logistic Regression: A Softer Decision Boundary

The problem with using a **hard threshold function** (like in the perceptron) is that it **cannot express uncertainty**. The function can only output **0 or 1**, meaning it makes **absolute decisions** — either "Yes" or "No", with no in-between.

---

####  The Problem

Such binary decisions work in some cases, but they are **too rigid** for many real-world situations. For example, if the model is only *slightly* confident about a prediction, it will still output 1 or 0 — with no way to indicate uncertainty.

---

####  The Solution: Logistic Function

A **logistic function** solves this by introducing a **soft threshold**. It maps any input into a **real number between 0 and 1**, representing the **confidence level** or **probability** of the prediction.

---

####  Example Interpretation

- A prediction of `0.91` could mean there's a **91% confidence** in the prediction being `1` (e.g., "Harry belongs to Gryffindor").
- A prediction of `0.52` means the model is **uncertain** and could lean either way.
- A prediction of `0.08` shows strong confidence


### Conclusion:
The result of the perceptron learning rule is a decision boundary that separates Gryffindor from Slytherin characters. If a new character’s courage and ambition values fall on one side of the decision boundary, they will be classified as Gryffindor; otherwise, they will be classified as Slytherin.

However, as real-world data is often noisy and imperfect, the perceptron may sometimes misclassify characters, especially when the features (Courage and Ambition) do not perfectly separate the two classes.


##  Support Vector Machines (SVM)

In addition to **nearest-neighbor** and **Perceptron Learning**, another powerful approach to classification is the **Support Vector Machine (SVM)**.

---

###  Core Idea

SVMs don't just aim to separate data correctly — they aim to do it **optimally**.

> They find the decision boundary (also known as a hyperplane) that maximizes the **margin** — the distance between the boundary and the nearest data points of each class.

These critical data points are called **Support Vectors**.

---

###  Why Margin Matters

All the decision boundaries might separate the data **without errors**, but not all are equally good.

- Boundaries **too close** to data points might be sensitive to noise and misclassify slightly different inputs.
- An SVM seeks the boundary that **maximizes the margin** between classes — the **Maximum Margin Separator**.

This makes the classifier more **robust** and **generalizable**.

---

###  Going Beyond Lines

SVMs can also:
- Handle data in **higher dimensions**.
- Model **non-linear** decision boundaries by using the **kernel trick**, such as the **Radial Basis Function (RBF)** kernel.

This allows SVMs to form **circular** or **curved boundaries** that separate data in a more flexible manner.

## Regression

**Regression** is a supervised learning task that maps an input point to a **continuous value**, i.e., some real number. This differs from classification, where the goal is to map an input to **discrete values** (e.g., *Gryffindor* or *Slytherin*).

For example, the publishers of *Harry Potter and the Order of the Phoenix* might use regression to estimate **how many copies** of the book will sell based on **how much effort** is put into promotion (ads, interviews, merchandise, etc.).

In this case, the observed function `f(promotion_effort)` represents the **actual number of copies sold** after a certain promotional budget was spent. These observed data points are what we start with.

The goal is to develop a **hypothesis function** `h(promotion_effort)` that can **predict future sales** based on how much promotion is done. This function `h` generates a **regression line**, which aims **not** to classify things into categories but to provide a **numerical estimate** for sales based on input values.

### Loss Functions

In the *Harry Potter and the Order of the Phoenix* book sales, the days where we predicted sales correctly are the days where the actual sales are close to the predicted values (sales that are near the regression line). However, days where the prediction was off — for example, we predicted more sales than actual, or vice versa — represent mistakes, where our predictions are further from the observed sales. These are the days where the prediction error is large.

We can assign a **loss** value to each day based on how far off our predictions were. Each mistake is quantified, with a value of 1 for each day we fail to predict correctly. The sum of these values gives an empirical estimate of how “bad” our regression line (or decision boundary) is.

```
# L₁: L(actual, predicted) = |actual - predicted|
```



- **L2 Loss (Absolute Loss)**: L₂ penalizes outliers more harshly than L₁ because it squares the the difference


## Overfitting in Machine Learning

Overfitting happens when a model performs **too well** on training data — so well that it memorizes it instead of learning general patterns. This results in poor performance on unseen or test data.

Think of it like this: the model becomes like a student who memorized all the questions from last year’s exam, but completely fails this year’s paper.

---

###  What Does Overfitting Look Like?

In machine learning, we use **loss functions** to guide the model during training. But loss functions can be a double-edged sword. Here's why:

- The model tries to **minimize the loss** on training data.
- If it's too focused on this, it can perfectly fit the noise and quirks of that dataset.
- This results in a **training loss ≈ 0**, but the model fails to perform well on **new, unseen data**.

---

###  Harry Potter Analogy

Let’s imagine a **Sorting Hat** model that predicts which house a student belongs to: Gryffindor, Slytherin, Ravenclaw, or Hufflepuff.

If the model overfits the training data:

- It may perfectly memorize that *"Harry with glasses = Gryffindor"* and *"Draco with blond hair = Slytherin"*.
- But when it sees *Luna Lovegood*, who doesn’t follow the usual patterns, it might wrongly assign her to **Gryffindor** just because she vaguely resembles a memorized Gryffindor feature.

Thus, the **sorting hat model fails to generalize**, even though it had near-perfect predictions on past students.

##  Regularization Overview

- **Loss** measures how far off predictions are.
- **Complexity** reflects the model's complexity (e.g., large weights in neural networks).
- **λ (lambda)** controls the penalty for complexity:
  - **λ = 0**: No penalty, the model can be very complex.
  - **λ too high**: Model might underfit due to simplicity.

---

###  Why Regularize?

Without regularization, your model might overfit, creating a complex curve to fit every point. Regularization encourages a simpler model that captures the general pattern, avoiding noise.

---

###  Detecting Overfitting

###  Holdout Cross-Validation:
- Split data into **training** and **test sets**.
- If the model performs well on training but poorly on testing, it's likely overfitting.

####  Limitation: Not all data can be used for training.

###  k-Fold Cross-Validation:
- Split data into **k** parts.
- Train on `k-1` folds, test on the remaining fold, and repeat **k times**.
- Average results for a better model evaluation.

---
###  Summary

- Regularization combines **loss** and **complexity penalty** to prevent overfitting.
- Controlled by **λ**, with higher values favoring simpler models.
- Use **cross-validation** to ensure the model generalizes well.

 **Regularization** is crucial in machine learning to keep your model from overfitting.

Now that we have our dataset, we can use scikit-learn to train a machine learning model to predict whether a new banknote is genuine or counterfeit.

Scikit-learn provides various models that can be used for this task. In this example, we will explore a few options:

````python
# Importing necessary libraries
import csv
import random

# Importing machine learning models from scikit-learn
from sklearn import svm
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

# Choose the model: Uncomment the one you want to use
# model = KNeighborsClassifier(n_neighbors=1)
# model = svm.SVC()
model = Perceptron()  # We're using Perceptron here for classification

# Mount Google Drive to access files
from google.colab import drive
drive.mount('/content/drive')

# Read data from the CSV file located in your Google Drive
csv_file_path = '/content/drive/MyDrive/datasets/banknotes.csv'  # Change this to the actual path to your CSV file
with open(csv_file_path) as f:
    reader = csv.reader(f)
    next(reader)  # Skip the header row

    data = []
    # Loop through each row of the dataset
    for row in reader:
        data.append({
            "evidence": [float(cell) for cell in row[:4]],  # The first 4 columns are features (evidence)
            "label": "Authentic" if row[4] == "0" else "Counterfeit"  # The last column is the label (0 or 1)
        })

# Split the dataset into training and testing sets
# 40% of the data is used for testing, the rest for training
holdout = int(0.40 * len(data))  # 40% for testing
random.shuffle(data)  # Shuffle the data randomly to ensure unbiased splitting
training = data[holdout:]  # 60% for training
testing = data[:holdout]  # 40% for testing


# Prepare the training data (X_training for features, y_training for labels)
X_training = [row["evidence"] for row in training]  # Features (evidence) for training
y_training = [row["label"] for row in training]  # Labels for training

# Train the model using the training data
model.fit(X_training, y_training)

# Prepare the testing data
X_testing = [row["evidence"] for row in testing]  # Features (evidence) for testing
y_testing = [row["label"] for row in testing]  # True labels for testing

# Make predictions using the trained model
predictions = model.predict(X_testing)

# Evaluate the model's performance
correct = 0  # Counter for correct predictions
incorrect = 0  # Counter for incorrect predictions
total = 0  # Total number of predictions made

# Compare the predicted results with the actual results
for actual, predicted in zip(y_testing, predictions):
    total += 1  # Increment the total number of predictions
    if actual == predicted:  # Check if prediction is correct
        correct += 1
    else:  # If the prediction is incorrect
        incorrect += 1

# Print the results
print(f"Results for model {type(model).__name__}")  # Display the model type (e.g., Perceptron)
print(f"Correct: {correct}")  # Number of correct predictions
print(f"Incorrect: {incorrect}")  # Number of incorrect predictions
print(f"Accuracy: {100 * correct / total:.2f}%")  # Calculate and print the accuracy
````

##  Reinforcement Learning

**Reinforcement Learning (RL)** is a machine learning approach where an agent learns to make decisions by interacting with an environment.

After each action, the agent receives feedback in the form of a **reward** (a positive or negative numerical value):

- **Positive rewards** reinforce the action (make it more likely to be repeated).
- **Negative rewards** (punishments) discourage the action.

###  How It Works

1. The **environment** provides a **state** to the agent.
2. The **agent** performs an **action**.
3. The **environment** returns:
   - A new **state**
   - A **reward** based on the action taken
4. The agent uses this reward to improve its future actions.








```
+-------------------+
|   Environment     |
+-------------------+
          |
          v
+-------------------+
|   Agent receives  |
|   initial state   |
+-------------------+
          |
          v
+-------------------+
|  Agent takes an   |
|      action       |
+-------------------+
          |
          v
+----------------------------+
| Environment returns        |
| new state + reward/punish  |
+----------------------------+
          |
          v
+----------------------------+
| Agent updates policy       |
| (learns from feedback)     |
+----------------------------+
          |
          v
+-----------------------------+
| Loop back to next action   |
+-----------------------------+
          |
          v
   (Repeat Until Done)

```



### 📌 Use Case Example

A common application is training **walking robots**:
- Taking a step forward: **+1 reward**

##  Markov Decision Processes (MDPs)

Reinforcement learning can be viewed as a **Markov Decision Process**, which includes the following properties:

- **Set of States (S)**  
  All the possible situations the agent can find itself in.

- **Set of Actions (Actions(S))**  
  A set of actions available to the agent from each state.

- **Transition Model (P(s’ | s, a))**  
  A probability model that determines the next state `s’` given the current state `s` and action `a`.

- **Reward Function (R(s, a, s’))**  
  A function that returns a numerical reward for transitioning from state `s` to state `s’` via action `a`.

---

###  Example

Imagine a simple grid world:

- The agent is represented by a **yellow circle**.
- The goal is to reach the **green square**.
- The **red squares** are dangerous and should be avoided.

Each square on the grid is a **state**.  
The agent can take **actions** like moving `up`, `down`, `left`, or `right`.

The **transition model** defines where the agent ends up after taking an action.  
The **reward function** defines the feedback:

- If the agent steps on a **red square**, it receives a **negative reward** (punishment).
- If it reaches the **green square**, it receives a **positive reward**.
- Otherwise, it receives **neutral or small negative rewards** to encourage quick solutions.

Over time, the agent **learns** which state-action combinations to avoid and which paths lead to high rewards.

---

###  Behavior Learning

Through trial and error, the agent:
- **Explores** the environment
- **Learns from rewards/punishments**
- **Adjusts** its action probabilities accordingly

This learning can be **probabilistic**, meaning the agent may explore alternate actions and increase the probability of successful ones while decreasing those of harmful actions.

---

###  Goal

When the agent finally reaches the green square and receives a **positive reward**, it reinforces the decision path that led to success — helping it repeat that behavior in future episodes.

###  Q-Learning: Reinforcement Learning in Action

Q-Learning is a **model-free reinforcement learning algorithm** where a function **Q(s, a)** estimates the value of taking action `a` in state `s`.

---

###  Initialization

- At the start, the agent **assumes zero knowledge**:

---

###  Learning through Feedback

When an action is taken and a **reward is received**, the Q-value gets updated using the following formula:

```
Q(s, a) ← Q(s, a) + α [r + γ * maxₐ′ Q(s′, a′) − Q(s, a)]
```

- **α (alpha)** is the learning rate:
  - α = 1 → new value completely overwrites old.
  - α = 0 → old value never updates.
  - 0 < α < 1 → gradual learning.
- **γ (gamma)** is the discount factor:

  - γ = 0 → the agent considers only immediate rewards.

  - γ = 1 → the agent considers long-term rewards fully.

  - 0 < γ < 1 → balances between short-term and long-term rewards.
---

###  Estimating the New Value

The new value estimate is computed by:



```
r + γ * maxₐ′ Q(s′, a′)
```




Where:
- `r` is the **immediate reward**.
- `γ` (gamma) is the **discount factor** for future rewards.
- `max_a' Q(s', a')` is the **maximum future reward** from the new state `s'`.

This update considers **not just the immediate reward**, but also the **long-term return**.

---

##  Unsupervised Learning

In all the cases we saw before, such as in **supervised learning**, we had **data with labels** that the algorithm could learn from.

> **Example:** When training a model to detect counterfeit notes, the input data consisted of numerical features, and the label was whether the note was counterfeit or not.

However, in **unsupervised learning**, we are only given **input data** without labels, and the AI's job is to identify patterns or groupings within the data itself.

---

###  Clustering

**Clustering** is an unsupervised learning task where the algorithm organizes the input data into groups. The goal is to have similar data points grouped together, while dissimilar data points are placed in different groups.

Clustering can be applied in various fields such as:
- **Genetics**: Identifying similar genes.
- **Image Segmentation**: Dividing an image into regions based on similarity of pixel values.

---

###  k-means Clustering

**k-means Clustering** is a popular algorithm used to perform clustering tasks. Here's how it works:

1. **Initialization**: The algorithm randomly selects **k** initial cluster centers (the number of clusters, **k**, is chosen by the programmer).
2. **Assignment**: Each data point is assigned to the cluster whose center is closest.
3. **Update**: The cluster centers are recalculated by moving them to the **median** of all points in that cluster.
4. **Repeat**: Steps 2 and 3 are repeated until the clusters stabilize, meaning that the assignments of points to clusters no longer change.

The algorithm stops when it reaches an equilibrium, with all points assigned to clusters that no longer change.

---

##  Core Truth About Unsupervised Learning

> **Most traditional unsupervised learning algorithms work best if they are given the correct number of clusters, *or* if they can estimate it accurately.**

###  Two Kinds of Algorithms:

| Algorithm Type                      | Needs You to Give K? | Notes |
|------------------------------------|-----------------------|-------|
| `KMeans`, `GMM`                    | ✅ Yes                | Fast, but assumes spherical clusters and same size. |
| `DBSCAN`, `HDBSCAN`, Agglomerative | ❌ No                 | Infers clusters based on density or structure. |

---

##  If You Give the Wrong K in KMeans

- **Too Few Clusters (Underfitting)** → Merges distinct groups.
- **Too Many Clusters (Overfitting)** → Splits real groups into nonsense.

---

##  Smarter Approaches Exist

### Techniques that **don't need K up front**:
- `DBSCAN` — density-based, detects noise too
- `HDBSCAN` — hierarchical version, more stable
- `Agglomerative` — you cut the dendrogram after building it

### Techniques that **help estimate K**:
- **Elbow Method**
- **Silhouette Score**
- **Gap Statistic**
- **AIC/BIC** (for GMM)
